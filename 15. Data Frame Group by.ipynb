- df['특정 컬럼'].unique() : 특정 컬럼의 유일한 값만을 표현. 보통 순서 척도 등 변수에 사용됨.
<br>
- df.groupby.통계함수() : 특정 그룹에 따른 통계 데이터를 보여줌.
    - 통계 함수 
        - sum()
        - mean() 
        - count() 
        - max() 
        - min() 
        - agg()

```python
# 데이터 읽기.
# index_col = '' : 인덱스로 설정해서 불러옴.
df = pd.read_excel('data/hotel.xlsx', index_col='Unnamed: 0')
df
```

```python
# 컬럼 'grade'의 유일한 값들만 표시.
df.grade.unique()
df['grade'].unique()
```

```python
# grade 값의 개수를 센 후 내림차순으로 정렬.
df['grade'].value_counts().sort_index()
```

```python
# 등급별 가격의 합( ~ 별 )
# DataFrameGroupBy(집계 관련함수 : sum, mean, count, max, min, agg)
df.groupby('grade').sum()[ ['price'] ]
```

```python
df.groupby('grade').min()[ ['price'] ]
```

```python
df.groupby('grade').count()[ ['price'] ]
```

```python
mydf = df.groupby('grade').agg(['sum', 'mean'])[['price']]

# 특정 컬럼명으로 정해줄 수 있음.
mydf.columns = ['가격의 총합', '가격의 평균']
mydf
```
