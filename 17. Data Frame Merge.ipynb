```python
import numpy as np
import pandas as pd
```

- merge() : 0데이터를 병합시켜줌. 단, 반드시 공통 컬럼이 존재해야 한다.
- concat() : 데이터를 병합시켜줌. 공통 컬럼이 존재하지 않아도 된다.
    - axis = 0 : 행 기준.
    - axis = 1 : 열 기준.

```python
# 데이터 만들기.
employee = [{'empno':1, 'ename':'kim', 'dept':1},
            {'empno':2, 'ename':'lee', 'dept':2},
            {'empno':3, 'ename':'park', 'dept':1},
            {'empno':4, 'ename':'song', 'dept':3},
            {'empno':5, 'ename':'min', 'dept':2}
           ]
dept = [{'dept':1, 'deptname':'관리직'},
        {'dept':2, 'deptname':'영업직'},
        {'dept':3, 'deptname':'개발직'}
       ]
info = [{'empno':1, 'addr':'서울시','phone':'010-1111-1111'},
       {'empno':3, 'addr':'부산시','phone':'010-2222-2222'},
       {'empno':2, 'addr':'광주시','phone':'010-3333-3333'},
       {'empno':5, 'addr':'광주시','phone':'010-4444-4444'},
       {'empno':4, 'addr':'광주시','phone':'010-5555-5555'}
       ]

# 데이터 프레임으로 만들기.
emp  = pd.DataFrame(employee)
dept = pd.DataFrame(dept)
info = pd.DataFrame(info)
```

```python
# on : 특정 컬럼('dept')을 기준으로 데이터를 병합.
mg1 = pd.merge(emp, dept, on='dept')

# dept 컬럼을 삭제.
mg1.drop(columns=['dept'], inplace=True)
mg1
```

```python
# on : 특정 컬럼('empno')을 기준으로 데이터를 병합.
mg2 = pd.merge(mg1, info, on='empno')

# empno 컬럼을 삭제.
mg2.drop(columns=['empno'], inplace=True)
mg2
```

```python
# concat() : 공통 컬럼이 존재하지 않아도 데이터 결합이 가능.
# default 값이 axis = 0 이므로 행으로 병합.
pd.concat([emp, dept])
```

```python
# axis=1 : 열로 병합.
pd.concat([emp, dept], axis=1)   
```
