```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 사용할 데이터 만들기.
data =  {'eng':[10,30,50,70],
         'kor':[20,40,60,80],
         'math':[90,50,20,70]}
df = pd.DataFrame(data, index = ['a','b','c','d'])
df
```

#### 시각화(Visualization)


```python
# line(선) 그래프.
# kind : 시각화 유형을 지정.
# style : 선 모양과 색상을 지정.
df.plot(kind='line', style=['ro--','b<-', 'ms-.'])
plt.show()
```

```python
df.plot(kind='bar', y=['eng','kor'], color=['coral', 'lightcoral'], alpha=0.7, grid=True, rot=45)
plt.show()
```

```python
# stacked=True
df.plot(kind='bar', stacked=True, color=['black', 'grey', 'dimgrey'], alpha=0.7, grid=True)
plt.text(0,50,'test1', va='center', ha='center')
plt.text(1,50,'test2')
plt.show()
```

```python
# Scatter(산점도)
df.plot(kind='scatter', x='eng' ,y='math', s=100, c='m', alpha=0.8)
plt.show()
```

```python
# Boxplot(상자 그림)
# 4분위수 및 이상치를 살펴볼 수 있음.
df.plot(kind='box')
plt.show()
```

```python
# 직접 숫자 위치를 맞춰가면 그림.
df.plot(kind='bar', stacked=True, color=['silver','gold', 'salmon'], alpha=0.7, grid=True)
plt.text(0,5,'10', va='center', ha='center')
plt.text(0,20,'20', va='center', ha='center')
plt.text(0,90,'90', va='center', ha='center')
plt.show()
```

```python
# 컬럼명만 추출.
for n in df: 
    print(n)
```

```python
# 인덱스 추출
for n in df.index: 
    print(n)
```

```python
# 값만을 추출.
for n in df.values: 
    print(n)
```

```python
# items() : 튜플로 추출
for n,sr, in df.items():
    print(n, sr['a'], sr['b'], sr['c'], sr['d'])
```

```python
# iterrows(): enumerate() 같은 개념으로 index랑 각각의 열을 추출.
for n, sr in df.iterrows():
    print(n,  sr['eng'], sr['kor'], sr['math'])
```

```python
df.plot(kind='bar', stacked=True, grid=True, alpha=0.7, color=['pink','skyblue', 'tomato'] )
n = 0

# iterrows(): enumerate() 같은 개념으로 index랑 각각의 열을 추출.
for p, sr in df.iterrows():
    #print(p,  sr['eng'], sr['kor'], sr['math'])
    e,k,m = (sr['eng'], sr['kor'], sr['math'])
    plt.text(n, e/2, f'{e}점', va='center', ha='center')
    plt.text(n, e+k/2, f'{k}점', va='center', ha='center')
    plt.text(n, e+k+m/2, f'{m}점', va='center', ha='center')
    n+=1
    
plt.show()
```
